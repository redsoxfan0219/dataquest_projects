## Removing Duplicates

Below I identify duplicates in the datasets and remove them thereafter.

```python
duplicate_android = []
unique_android = []
duplicate_ios = []
unique_ios = []

for app in android:
    android_name = app[0]
    if android_name in unique_android:
        duplicate_android.append(android_name)
    else: 
        unique_android.append(android_name)
print('Number of duplciate Android apps:', len(duplicate_android))

for app in ios:
    ios_name = app[0]
    if ios_name in unique_ios:
        duplicate_ios.append(ios_name)
    else: 
        unique_ios.append(ios_name)
print('Number of duplicate iOS apps:', len(duplicate_ios))

print('Sample duplicates', duplicate_android[1:5])
```

Duplicates in the Android dataset appear to differ only in the number of reviews they received, implying that they were collected at different times. I will use that information as the basis for removing duplicates, removing those duplicates with fewer reviews. After removing the duplicates, I would expect the cleaned dataset to be 9659 rows:

```python
print('Expected length:', len(android)-1181)
```

```python
reviews_max = {}
for app in android:
    name = app[0]
    n_reviews = float(app[3])
    
    if name in reviews_max and reviews_max[name] < n_reviews:
        reviews_max[name] = n_reviews
        
    elif name not in reviews_max:
        reviews_max[name] = n_reviews

print(len(reviews_max))
```

Above, I created a dictionary that updates a key-value pair of app name and number of reviews, updating the number of reviews when a duplicate app row was encountered. I then print the length of the dictionary to compare against the expected length calculated above. 

Below, I use that newly created dictionary to remove duplicate rows from the original dataset. The final result is the duplicate-less dataset, of which I print the first ten rows for illustration.

## Removing Non-English App Data

Next, I'll work to remove the non-English language apps from the dataset. To do so, I will remove rows of data with names that fall outside the ASCII range that represents English-language characters (represented as numbers 0-127). I will loop over each character in the title of an app name to determine whether it falls outside the English-language range, and then remove rows with names that have characters falling outside that range. So as not to exclude English-language apps which have the occasional non-standard symbol (e.g., the TM symbol), I will only exclude apps that have more than 3 symbols outside non-standard English characters.

```python
def noneng(string):
    non_ascii = 0 
    for character in string:
        if ord(character) > 127:
            non_ascii += 1
    if non_ascii > 3:
        return False
    else:
        return True
    
```

## Frequency Tables for App Profile

Our goal is to initially build our app for Android. If we determine that it is profitable after six months, we will further develop it, including building out an iOS version. Thus, we need to determine attributes of a successful app in both markets. Below, I build out frequency tables for both markets to determine the attributes of a successful app.

Having studied the column names, I believe building frequency tables for price, primary genre / Category, and average rating will help me determine successful attributes.

Below, I use two functions to build frequency tables and order the tables so that they appear in order of descending percentages.

```python
def freq_table(dataset, index):
    freq ={}
    total = 0
    for row in dataset:
        total += 1
        value = row[index]
        if value in freq:
            freq[value] += 1
        else:
            freq[value] = 1
            
    table_percentages ={}
    for key in freq:
        percentage = (freq[key] / total)*100
        table_percentages[key] = percentage
        
    return table_percentages

def display_table(dataset, index):
    table = freq_table(dataset, index)
    table_display = []
    for key in table:
        key_val_as_tuple = (table[key], key)
        table_display.append(key_val_as_tuple)

    table_sorted = sorted(table_display, reverse = True)
    for entry in table_sorted:
        print(entry[1], ':', entry[0])
```
