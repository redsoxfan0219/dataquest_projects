## Exploring Unuseful Information and Columns to Be Revised
Next, I will be looking for unuseful information to drop from the dataset and other columns that contain numeric information currently in string format.

```python
autos.describe(include='all')
```

As noted earlier in this document, 'price' and 'odometer' need to be converted to int. I do this below. 'seller' and 'offer_type' are largely without variance, so I will remove those columns from the dataset.

```python
cleaned_price = []
cleaned_odometer = []

for price in autos['price']:
    price = price.replace("$", "")
    price = price.replace(",", "")
    price = int(price)
    cleaned_price.append(price)
    
for odo in autos['odometer']:
    odo = odo.replace("km", "")
    odo = odo.replace(",", "")
    odo = int(odo)
    cleaned_odometer.append(odo)
    
autos['price'] = cleaned_price
autos['odometer'] = cleaned_odometer
   
autos.rename(columns = {"odometer":"odometer_km"}, inplace = True)
```

Given the diversity of brands, I am going to look at the top 10 manufactuers and the average price associated with each.

```python
brands = ['volkswagen', 'bmw','opel', 'mercedes_benz', 'audi', 'ford', 'renault', 'peugeot', 'fiat', 'seat']
top_10_avg_price = {}
for brand in brands:
    brand_bool = autos['brand'] == brand
    brand_slice = autos[brand_bool]
    brand_avg_price = brand_slice['price'].mean()
    top_10_avg_price[brand] = int(round(brand_avg_price))
    
print(top_10_avg_price)
        
```

Based on the results above, we can conclude that there is a relatively surprising range of prices in the top selling manufactuers. Opel was on the cheaper end at approximately 3000 per auto, whereas BMW averaged more than 8000 and Audi averaged more than 9000. Volkswagen, the top seller, was right in the middle, at approximately 5400 on average. 

Now I'll add average mileage per sale to help determine how this affects the sale price. I'll then create a new dataframe with the mean price and mean mileage for each of the top brands.

```python
top_10_avg_odo = {}
for brand in brands:
    brand_bool = autos['brand'] == brand
    brand_slice = autos[brand_bool]
    brand_avg_odo = brand_slice['odometer_km'].mean()
    top_10_avg_odo[brand] = int(round(brand_avg_odo))
    
print(top_10_avg_odo)
```

```python
avg_price_series = pd.Series(top_10_avg_price)
avg_odo_series = pd.Series(top_10_avg_odo)
compare = pd.DataFrame(avg_price_series,columns =['mean_price'])
compare['mean_odometer_km'] = avg_odo_series
print(compare)
```
