```python
import pandas as pd
import numpy as np
```

```python
# Read in Data
star_wars = pd.read_csv("star_wars.csv", encoding="ISO-8859-1")
```

```python
star_wars.columns
```

## Converting to Boolean

```python
seen = star_wars["Have you seen any of the 6 films in the Star Wars franchise?"]
fans = star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"]

yes_no = {
    "Yes": True,
    "No": False
}

seen = seen.map(yes_no)
fans = fans.map(yes_no)

star_wars["Have you seen any of the 6 films in the Star Wars franchise?"] = seen
star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"] = fans
star_wars.head()
```

```python
# Convert to Boolean Values in the Columns
mapping = {
    "Star Wars: Episode I  The Phantom Menace": True,
    "Star Wars: Episode II  Attack of the Clones":True,
    "Star Wars: Episode III  Revenge of the Sith":True,
    "Star Wars: Episode IV  A New Hope":True,
    "Star Wars: Episode V The Empire Strikes Back":True,
    "Star Wars: Episode VI Return of the Jedi":True,   
    np.NaN: False}

for col in star_wars.columns[3:9]:
    star_wars[col]=star_wars[col].map(mapping)

star_wars
```

```python
# Renaming Columns

star_wars = star_wars.rename(columns = {
                             "Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1",
                             "Unnamed: 4": "seen_2",
                             "Unnamed: 5": "seen_3",
                             "Unnamed: 6": "seen_4",
                             "Unnamed: 7": "seen_5",
                             "Unnamed: 8": "seen_6",
})
star_wars.head()
                             
```

```python
# Convert Ranking Columns to Floats

star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
```

```python
# Change Names of Ranking Columns

star_wars = star_wars.rename(columns ={
    "Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.":"rank_1",
    "Unnamed: 10":"rank_2",
    "Unnamed: 11":"rank_3",
    "Unnamed: 12":"rank_4",
    "Unnamed: 13":"rank_5",
    "Unnamed: 14":"rank_6",
})
star_wars.columns
```

## Finding the Mean Rankings

```python
for col in star_wars.columns[9:15]:
    mean = star_wars[col].mean()
    print(f"The average ranking {col} is: {mean}")
```

- Based on the average rank, it seems there is a clear preference for viewing Episode V as the best of the series.

## Number of Individuals Who Have Seen Each Film

```python
for col in star_wars.columns[3:9]:
    total = star_wars[col].sum()
    print(total)
```

- I'm guessing the viewship happened the way it did due to popularity driven by word of mouth. I remember people being very excited by the prospect of Episode I when it came out, but then there was a lot of frustration because this movie wasn't great. The trend continued for Eps II and III. 

## Breaking Down Ranks by Fandom and Gender

```python
males = star_wars[star_wars["Gender"] == "Male"]
females = star_wars[star_wars["Gender"] == "Female"]
fans = star_wars[star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"] == True]

for col in males.columns[9:15]:
    mean = males[col].mean()
    print(f"The average male ranking for {col} is: {mean}")

print("")

for col in females.columns[9:15]:
    mean = females[col].mean()
    print(f"The average female ranking for {col} is: {mean}")
    
print("")

for col in fans.columns[9:15]:
    mean = fans[col].mean()
    print(f"The average fan ranking for {col} is: {mean}")
                 
```

- Across the board, everyone likes Episode V the best. Males like Episode IV slightly more than females and fans, who generally think Episode VI is bette than IV. There is pretty universal agreement that Ep III is the worst, but this is also partly influenced by the fact that fewer people watched that movie.
